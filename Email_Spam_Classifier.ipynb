```python
import pandas as pd
```

```python
messages = pd.read_csv("/content/spam (1).csv")
```

```python
messages
```

Data Cleaning and Preprocessing

```python
import nltk
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
import re
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download('punkt_tab')
```

```python
wordnet = WordNetLemmatizer()
corpus = []
```

```python
for i in range(0,len(messages)):
  review = re.sub("[^a-zA-z]" , " " , messages["Message"][i])
  review  = review.lower()
  review = review.split()
  review = [wordnet.lemmatize(word) for word in review if word not in set(stopwords.words("english"))]
  review = " ".join(review)
  corpus.append(review)
```

Creating the TF - IDF model

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=1500)
X = tfidf.fit_transform(corpus).toarray()
y = pd.get_dummies(messages['Category'])
y = y.iloc[:,1].values
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.2)
```

Training the model with Naive Bayes Algorithm

```python
from sklearn.naive_bayes import MultinomialNB
email_spam = MultinomialNB()
email_spam.fit(X_train , y_train)
y_pred = email_spam.predict(X_test)
```

Checking the Accuracy of the model

```python
from sklearn.metrics import confusion_matrix , accuracy_score
print(accuracy_score(y_test , y_pred))
cmax = confusion_matrix(y_test , y_pred)
Confusion_matrix = pd.DataFrame(cmax)
Confusion_matrix
```
